Questão: Quantos pessoas com o nome jhon morreram?

```python
import pandas as pd

# Carregar o dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv")

# Filtrar por linhas onde o campo "Name" contém "Jhon" (ignorando maiúsculas/minúsculas)
filtro = df['Name'].str.contains('john', case=False, na=False)

# Contar quantas dessas pessoas morreram (Survived == 1)
mortes = df.loc[filtro & (df['Survived'] == 0)]

# Exibir o total de mortes
print(f"Quantidade de pessoas chamadas 'Jhon' que morreram: {len(mortes)}")
```

 Este código categoriza as idades dos passageiros em faixas etárias e calcula a taxa de sobrevivência média para cada faixa etária, permitindo analisar como a idade influenciou a sobrevivência no Titanic.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv")

df['idade_faixa'] = pd.cut(df['Age'], bins=[0, 18, 30, 45, 60, 80], include_lowest=False)
df[['idade_faixa', 'Survived']].groupby('idade_faixa').mean()
```

Este código categoriza as idades dos passageiros em faixas etárias e, para cada faixa, calcula a taxa de sobrevivência média e o número total de passageiros. Isso fornece uma visão detalhada de como a sobrevivência varia entre diferentes grupos etários e quantos passageiros estão em cada grupo.



```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv")

df['idade_faixa'] = pd.cut(df['Age'], bins=[0, 18, 30, 45, 60, 80], include_lowest=False)
df['one'] = 1
df[['idade_faixa', 'Survived', 'one']].groupby('idade_faixa').agg({'Survived':['mean'],'one':['sum']})
```
